```python
import heapq 
def a_star_short(grid, start, end): 
    queue, g_score = [(0, start)], {start: 0} 
    parents = {}  
 
    while queue: 
        _,curr = heapq.heappop(queue) 
        if curr == end: break 
 
        for r, c in [(curr[0]+x, curr[1]+y) for x, y in [(-1,0), (1,0), (0,-1), (0,1)]]: 
            if 0 <= r < len(grid) and 0 <= c < len(grid[0]) and grid[r][c] == 0: 
                new_cost = g_score[curr] + 1 
                if new_cost < g_score.get((r, c), float('inf')): 
                    g_score[(r, c)], parents[(r, c)] = new_cost, curr 
                    heapq.heappush(queue, (new_cost + abs(end[0]-r) + abs(end[1]-c), (r, c))) 
 
    path, node = [], end 
    while node in parents: path.append(node); node = parents[node] 
    return [start] + path[::-1] if path else [] 
 
maze = [ 
    [0, 0, 0, 1, 0], 
    [0, 1, 0, 1, 0], 
    [0, 1, 0, 0, 0] 
] 
print(f"Path: {a_star_short(maze, (0, 0), (2, 4))}")
```
