```python
from collections import deque

graphs = {
    "A":["B","C"],
    "B":["D","E"],
    "C":["F"],
    "D":[],
    "E":["F"],
    "F":[]
}

def bfs(graph,start,goal):
    visited = set()
    queue = deque([[start]])
    
    while queue:
        path = queue.popleft()
        node = path[-1]
        if node == goal:
            return path
        if node not in visited:
            for neighbor in graph[node]:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
            visited.add(node)
    return None

def dfs(graphs,start,goal):
    stack = [[start]]
    visited = set()
    while stack:
        path = stack.pop()
        node = path[-1]
        if node == goal:
            return path
        if node not in visited:
            for neighbor in graphs[node]:
                new_path = list(path)
                new_path.append(neighbor)
                stack.append(new_path)
            visited.add(node)
    return None 

start_node = "A"
goal_node = "F"

bfs_path = bfs(graphs,start_node,goal_node)
dfs_path = dfs(graphs,start_node,goal_node)

print(f"Graph: {graphs}")
print(f"\n finding path from {start_node} to {goal_node}")
print("\n BFS path(Shortest path):")
print("->".join(bfs_path) if bfs_path else "No path found")
print("\n DFS path:")
print("->".join(dfs_path) if dfs_path else "No path found")
```
