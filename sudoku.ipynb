```python
board = [ 
 [5,3,0,0,7,0,0,0,0], 
 [6,0,0,1,9,5,0,0,0], 
 [0,9,8,0,0,0,0,6,0], 
 [8,0,0,0,6,0,0,0,3], 
 [4,0,0,8,0,3,0,0,1], 
 [7,0,0,0,2,0,0,0,6], 
 [0,6,0,0,0,0,2,8,0], 
 [0,0,0,4,1,9,0,0,5], 
 [0,0,0,0,8,0,0,7,9] 
] 
 
def solve(b): 
    # Find the first empty cell (row, col), or return True if none found 
    find = next(((r, c) for r in range(9) for c in range(9) if b[r][c] == 0), None) 
    if not find: return True 
    r, c = find 
 
    for n in range(1, 10): 
        # Check Row, Column, and 3x3 Box validation in one conditional block 
        if (n not in b[r]) and (n not in [b[i][c] for i in range(9)]) and (n not in [b[r//3*3+i][c//3*3+j] for i in range(3) for j in range(3)]): 
            b[r][c] = n 
            if solve(b): return True # Recursion 
            b[r][c] = 0 # Backtrack 
    return False 
 
# Execution 
if solve(board): 
    print("   Solved Sudoku:"); print(*board, sep='\n') 
else: 
    print("  No solution found")
```
